# **I.Xây dựng hàm phân tích**

```python
def analyze_and_plot(df, columns_to_check, new_column_names, group):
  symbols = ['①', '②', '③', '④', '⑤']
  symbol_labels = ['Hoàn toàn không đồng ý', 'Không đồng ý', 'Bình thường', 'Đồng ý', 'Rất đồng ý']
  # Create a dictionary to store the results
  results = {}

  # Calculate symbol counts for each column
  for column in columns_to_check:
      results[new_column_names[column]] = {symbol: df[column].str.contains(symbol, na=False).sum() for symbol in symbols}

  # Calculate "No Opinion" count for each column
      total_rows = len(df)
      symbols_count = sum(results[new_column_names[column]].values())
      results[new_column_names[column]]["Không có ý kiến"] = total_rows - 1 - symbols_count

  # Create a DataFrame from the results
  result_df = pd.DataFrame(results).T.fillna(0).astype(int)

  # Group by the first part of the column names (A1, A2, etc.) and sum
  grouped_results = result_df.groupby(result_df.index).sum()

  # Sau khi tính toán grouped_results:
  # Chuyển đổi bảng thành HTML với thẻ style để thêm đường kẻ
  table_html = tabulate(grouped_results, headers='keys', tablefmt='html')
  styled_table_html = f"""
  <style>
  table {{
      border-collapse: collapse; /* Gộp các đường viền của các ô */
      font-size: 16px; /* Kích thước chữ */
  }}
  table, th, td {{
      border: 1px solid black; /* Thêm đường viền cho bảng, tiêu đề và ô dữ liệu */
  }}
  </style>
  {table_html}
  """


  # Thêm cột "Không có ý kiến" vào dữ liệu cho biểu đồ
  symbol_counts = grouped_results[symbols + ["Không có ý kiến"]].sum()  # Lấy tổng cho các cột symbol và "Không có ý kiến"

  # Tính phần trăm cho mỗi symbol (bao gồm "Không có ý kiến")
  symbol_percentages = (symbol_counts / symbol_counts.sum()) * 100

  # Tạo tiêu đề cho bảng
  table_title = f"<h3>Kết quả khảo sát nhóm {group}</h3>"

  # Hiển thị tiêu đề và bảng HTML trong Jupyter Notebook
  display(HTML(table_title + styled_table_html))

  # Thêm đường kẻ ngang
  display(HTML('<hr>'))  # <hr> là thẻ HTML tạo đường kẻ ngang

  # Vẽ biểu đồ tròn
  plt.figure(figsize=(8, 8))  # Điều chỉnh kích thước biểu đồ
  wedges, texts, autotexts = plt.pie(symbol_percentages, autopct='%1.1f%%', startangle=90)
  # wedges: Danh sách các wedge (miếng) của biểu đồ tròn
  # texts: Danh sách các đối tượng text (phần trăm) trên biểu đồ tròn
  # autotexts: Danh sách các đối tượng text (nhãn) trên biểu đồ tròn

  plt.title('Phần trăm mức độ đồng ý của nhóm' + group)
  plt.axis('equal')  # Đảm bảo biểu đồ tròn có hình tròn

  labels = []
  for i, (l, s) in enumerate(zip(symbol_counts.index, symbol_percentages)):
    if l in symbols:
      labels.append(f'{l} - {symbol_labels[symbols.index(l)]} ({s:.1f}%)')
    else:
      labels.append(f'{l} ({s:.1f}%)')  # For "Không có ý kiến"

  plt.legend(wedges, labels, title="Chú thích", loc="center left", bbox_to_anchor=(1, 0, 0.5, 1))
  plt.show()



```
